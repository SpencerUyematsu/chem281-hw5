```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv("data/results.csv")
results
```

```python
plt.figure(figsize=(10, 10))

sns.lineplot(data=results,x="R", y="state_0", label="state 0")
sns.lineplot(data=results,x="R", y="state_1", label="state 1")
sns.lineplot(data=results,x="R", y="state_2", label="state 2")
sns.lineplot(data=results,x="R", y="state_3", label="state 3")
sns.lineplot(data=results,x="R", y="state_4", label="state 4")
sns.lineplot(data=results,x="R", y="state_5", label="state 5")
sns.lineplot(data=results,x="R", y="HF", label="HF")

plt.xlabel("Radius ($\AA$)")
plt.ylabel("Energy")
```
